# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

# Importing the dataset

```python
df = pd.read_csv("C:/Users/Dell/Desktop/datasets/Restaurant_Reviews.tsv", delimiter = '\t', quoting = 3)
df.head()
```

```python
corpus = []
for i in range(0, 1000):
    review = re.sub('[^a-zA-Z]', ' ', df['Review'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, 1000):
    review = re.sub('[^a-zA-Z]', ' ', df['Review'][i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

# Creating a bag-of-words model

```python
tf = TfidfVectorizer(max_features = 1500)
X = tf.fit_transform(corpus).toarray()
y = df.iloc[:, 1].values
```

```python
X
```

```python
y
```

# Splitting into Training and Test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Naive Bayes Model

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

# Predicting the test result

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

# Compare Actual and Predicted values

```python
compare = pd.DataFrame({'Actual': y_test, 'Predict': y_pred})
```

# Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print("accuracy for test data:",accuracy_score(y_test, y_pred))
```

```python
print("accuracy for train data:", classifier.score(X_train, y_train))
```

```python
print("classification_report:", classification_report(y_test, y_test))
```

```python
print("confusion_matrix:", confusion_matrix(y_test, y_pred))
```
