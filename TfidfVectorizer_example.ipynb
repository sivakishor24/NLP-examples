```python
paragraph = "Nature Nature is the most commonly used term.The word is used in various contexts. The most important indication is that the land is rich from multi-species plants, animals, wildlife and topography such as mountains, valleys, beaches and seas. , And forests. Natural beauty Man's nature is characterized by its beauty, which is a wonderful diversity of creatures in different parts of the earth, as well as the unique landscape of mountains, water, plateaus and forests. Each of these features is distinguished from other territories by its unique beauty, and this great diversity gives people a wide range of tranquility, tranquility, and tranquility. Most expressions of natural beauty are natural expressions for nature. Wherever a person turns his face he inevitably finds the manifestation of these. Perhaps the most striking of these is the unique variation in the colors around us, which creates a stunning natural painting. Other manifestations include sizes, shapes, and variations even in organisms of the same species. On the other hand, one of the greatest expressions of natural beauty is the beautiful aromas of certain organisms, such as flowers, flowers and certain kinds of fruits. , Especially citrus, which is composed of seasons and sometimes; Like the smell of earth after rain."
```

```python
#cleaning the texts

import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
#stemming
ps = PorterStemmer()
sentence = nltk.sent_tokenize(paragraph)
corpus = []
for i in range(len(sentence)):
    review = re.sub('[^a-zA-Z]', ' ', sentence[i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#lemmatization
wordnet = WordNetLemmatizer()
sentence = nltk.sent_tokenize(paragraph)
sentence1 = nltk.word_tokenize(paragraph)
corpus = []
for i in range(len(sentence)):
    review = re.sub('[^a-zA-Z]', ' ', sentence[i])
    review = review.lower()
    review = review.split()
    review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
#Count frequencies
frequency_list = []
import pprint
from collections import Counter

frequency_list = [Counter(d) for d in sentence1]
pprint.pprint(frequency_list)
```

```python
#creating the bag of words model
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
```

```python
cv.fit(corpus)
cv.get_feature_names()
```

```python
print("vocabulary content:\n {}".format(cv.vocabulary_))
```

```python
bag_of_words=cv.transform(corpus)
print(bag_of_words)
```
